```python
from pyspark import SparkContext
sc = SparkContext("local", "First App")
```

```python
file = sc.textFile("file_with_blank_lines.dat")
blankLines = sc.accumulator(0)
print file.collect()

def countBlankLines(line):
    global blankLines 
    
    if (line == ""):
       blankLines += 1
    
    return line.split(" ")
    
lines = file.flatMap(countBlankLines)
lines.coalesce(1).collect()

print "Blank Lines: %s" % blankLines.value    
```
